```python
# 문자열들이 들어있는 리스트
# 각 요소는 단어 하나씩이다.
a = ['alpha', 'bravo', 'charlie', 'delta', 'echo', 'foxtrot', 'golf', 'hotel', 'india']


# --------------------------------------------------
# enumerate(a) 확인용 코드
# --------------------------------------------------

# enumerate(a)는 리스트 a를 반복하면서
# 각 요소의 "인덱스"와 "값"을 함께 꺼내준다.
#
# 예:
# 0 alpha
# 1 bravo
# 2 charlie
# ...
#
# 여기서 index는 인덱스 번호,
# value는 해당 인덱스에 들어있는 실제 값이다.
for index, value in enumerate(a):
    print(index, value)


# --------------------------------------------------
# 글자 수가 5인 단어만 찾기
# --------------------------------------------------

# b라는 새 리스트를 만든다.
#
# 리스트 컴프리헨션 구조:
# [리스트에_넣을_값 for 변수 in 반복대상 if 조건]
#
# 여기서는 enumerate(a)를 사용하므로
# 반복할 때마다 index와 value를 함께 꺼낸다.
#
# enumerate(a)가 주는 값:
# (0, 'alpha')
# (1, 'bravo')
# (2, 'charlie')
# ...
#
# for index, value in enumerate(a)
# → index에는 인덱스가 들어감
# → value에는 실제 단어가 들어감
#
# if len(value) == 5
# → value의 글자 수가 5개인 경우만 통과
#
# (index, value)
# → 조건을 통과한 값은 인덱스와 단어를 튜플로 묶어서 리스트에 넣음
# b = [(index, value) for index, value in enumerate(a) if len(value) == 5]
b = [value for index, value in enumerate(a) if len(value) == 5]

# 최종 결과 출력
print(b)
```
