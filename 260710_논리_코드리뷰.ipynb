1. 나이가 20이상이면 성인, 미만이면 청소년으로 처리
2.  성인 남성입니다., 성인 여성입니다., 청소년 남성입니다. 청소년 여성입니다.
3. 남,여를 잘못 입력했을 시 '다시 입력해 주세요. [남, 여]'
4. 남 여를 판단, 이후 나이 판단해서 결과는 두개가 조인함수 써서 나오게 해보자. 이건 케이스를 다 써서 너무 길어.
    - age 먼저 물어봐서 if 쓰고 성인, 청소년 여부 그 하위에 남,여 물어봐서 결과를 합치게(합치는건 어떤 방법 써야할까)
    - 따로 따로 적용된 조건이 합쳐지는 방법이 있을까? 나중에 다 따로해서 별로인가?

```python
age =int(input('나이: '))
gender = input('성별[남,여]:')

if gender == '남':
    if age >= 20:# pass # 이거뭐임 코드를 우선 오류가 안나오게 넘기는 것
        print('성인 남성입니다.')
    else:
        print('청소년 남성입니다.')

elif gender =='여' :
    if age >=20:
        print('성인 여성입니다.')
    else:
        print('청소년 여성입니다.')
else:
    print('다시 입력해 주세요.[남, 여]') 
```

```python
# 사용자에게 나이를 입력받는다.
# input()은 기본적으로 문자열(str)로 입력을 받기 때문에,
# 나이 비교를 하려면 int()로 숫자형으로 바꿔야 한다.
age = int(input('나이: '))

# 사용자에게 성별을 입력받는다.
# 여기서는 '남' 또는 '여'만 정상 입력으로 처리한다.
gender = input('성별[남,여]: ')


# -------------------------------
# 1. 성별 판단
# -------------------------------

# 만약 gender에 들어온 값이 '여'라면
if gender == '여':
    # 출력에 사용할 성별 문자를 '여성'으로 저장한다.
    gender_text = '여성'

# 만약 gender에 들어온 값이 '남'이라면
elif gender == '남':
    # 출력에 사용할 성별 문자를 '남성'으로 저장한다.
    gender_text = '남성'

# '남'도 아니고 '여'도 아니라면 잘못 입력한 것이다.
else:
    # 잘못 입력했다는 안내 문구를 출력한다.
    print('다시 입력해 주세요. [남, 여]')

    # 성별이 잘못 입력되면 gender_text를 만들 수 없기 때문에,
    # 여기서 프로그램을 종료한다.
    # exit()가 없으면 아래쪽 print에서 gender_text가 없어서 오류가 날 수 있다.
    exit()


# -------------------------------
# 2. 나이 판단
# -------------------------------

# 조건에서 "20세 이상이면 성인"이라고 했으므로
# age >= 20 으로 비교한다.
if age >= 20:
    # 출력에 사용할 나이 구분 문자를 '성인'으로 저장한다.
    age_text = '성인'

# 20세 미만이면 청소년이다.
else:
    # 출력에 사용할 나이 구분 문자를 '청소년'으로 저장한다.
    age_text = '청소년'


# -------------------------------
# 3. 최종 출력
# -------------------------------

# f-string 문법을 사용해서 age_text와 gender_text를 문장 안에 넣는다.
# 예를 들어 age_text = '성인', gender_text = '남성'이면
# 최종적으로 '성인 남성입니다.'가 출력된다.
print(f'{age_text} {gender_text}입니다.')
```
